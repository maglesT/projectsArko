```python
import cv2
import os
import numpy as np
import glob
from PIL import Image

def convert_mask_to_bboxes(mask_path, class_id=0):
    """
    Convert a segmentation mask to YOLO-compatible bounding boxes.
    """
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    bboxes = []
    h, w = mask.shape
    for cnt in contours:
        x, y, w_box, h_box = cv2.boundingRect(cnt)
        x_center = (x + w_box / 2) / w
        y_center = (y + h_box / 2) / h
        width = w_box / w
        height = h_box / h
        bboxes.append(f"{class_id} {x_center} {y_center} {width} {height}")
    return bboxes

# Paths to images and masks
images = sorted(glob.glob('/content/drive/MyDrive/imss/Image/*.jpg'))
masks = sorted(glob.glob('/content/drive/MyDrive/imss/Mask/*.png'))

# Output directory for YOLO dataset
output_dir = "/content/yolo_dataset"
os.makedirs(f"{output_dir}/images/train", exist_ok=True)
os.makedirs(f"{output_dir}/labels/train", exist_ok=True)

# Re-save images using Pillow and prepare YOLO annotations
for img_path, mask_path in zip(images, masks):
    try:
        # Re-save the image using Pillow
        img = Image.open(img_path)
        img = img.convert('RGB')  # Ensure RGB format
        image_name = os.path.basename(img_path)
        img.save(f"{output_dir}/images/train/{image_name}", format='JPEG')  # Save in YOLO folder

        # Generate bounding box annotations
        bboxes = convert_mask_to_bboxes(mask_path)
        label_name = os.path.splitext(image_name)[0] + ".txt"
        with open(f"{output_dir}/labels/train/{label_name}", "w") as f:
            f.write("\n".join(bboxes))

        print(f"Processed: {img_path} and {mask_path}")
    except Exception as e:
        print(f"Error processing {img_path} or {mask_path}: {e}")
```

```python
import os
import shutil
import glob
from sklearn.model_selection import train_test_split

# Paths
image_dir = "/content/yolo_dataset/images/train"
label_dir = "/content/yolo_dataset/labels/train"

# Create directories for train and validation
os.makedirs("/content/yolo_dataset/images/val", exist_ok=True)
os.makedirs("/content/yolo_dataset/labels/val", exist_ok=True)

# List all images and labels
images = sorted(glob.glob(f"{image_dir}/*.jpg"))
labels = sorted(glob.glob(f"{label_dir}/*.txt"))

# Split into train and validation (80% train, 20% val)
train_images, val_images, train_labels, val_labels = train_test_split(
    images, labels, test_size=0.2, random_state=42
)

# Move validation images and labels to 'val' directories
for img_path, lbl_path in zip(val_images, val_labels):
    shutil.move(img_path, "/content/yolo_dataset/images/val")
    shutil.move(lbl_path, "/content/yolo_dataset/labels/val")

print("Train and validation data split completed!")

```
